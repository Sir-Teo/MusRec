```python
from bing_image_downloader import downloader

music_genre = {"country", "pop", "rock", "jazz", "blues", "metal", "hiphop", "classical", "reggae", "disco", "funk", "punk", "soul", "rnb", "indie", "electronic", "folk", "alternative", "dance"}
for genre in music_genre:
    downloader.download(genre, limit=50,  output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
```
