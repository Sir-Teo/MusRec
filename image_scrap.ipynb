```python
from bing_image_downloader import downloader

music_genre = {"country", "pop", "rock", "jazz", "blues", "classical", "reggae", "punk", "soul", "electronic", "folk"}
for genre in music_genre:
    downloader.download(genre, limit=300,  output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
```

```python
import os
for genre in music_genre:
    files_in_dir = os.listdir('dataset/'+genre)
    for file in files_in_dir:
        if not file.endswith('.jpg'):
            os.remove('dataset/soul/' + file)
```
