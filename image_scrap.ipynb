```python
from bing_image_downloader import downloader

music_genre = {"country music", "pop music", "rock music", "jazz music", "classical music"}
for genre in music_genre:
    downloader.download(genre, limit=50,  output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
```

```python
import os
music_genre = {"country music", "pop music", "rock music", "jazz music", "classical music"}
for genre in music_genre:
    files_in_dir = os.listdir('dataset/'+genre)
    for file in files_in_dir:
        if not (file.endswith('.jpg') or file.endswith('.jpeg')):
            os.remove('dataset/'+genre+'/' + file)
            break
        new_file = file.replace('.jpg', '.jpeg')
        os.rename('dataset/'+genre+'/' + file, 'dataset/'+genre+'/' + new_file)
```
